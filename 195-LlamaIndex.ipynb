```python
import requests
import os

pdf_url = 'https://raw.githubusercontent.com/JSJeong-me/GPT-Web/main/images/%EC%B9%B4%EC%B9%B4%EC%98%A4%ED%8E%98%EC%9D%B4-%EA%B1%B0%EB%9E%98%ED%99%95%EC%9D%B8%EC%A6%9D-20230700.pdf'
output_dir = './data'
output_filename = os.path.join(output_dir, '카카오페이-거래확인증-20230700.pdf')

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Download the PDF file
# Use stream=True for large files and iterate over content
response = requests.get(pdf_url, stream=True)
response.raise_for_status() # Raise an exception for HTTP errors

with open(output_filename, 'wb') as pdf_file:
    for chunk in response.iter_content(chunk_size=8192):
        pdf_file.write(chunk)

print(f"PDF downloaded successfully to {output_filename}")
```

```python
import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from google.colab import userdata

OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

# Set the OpenAI API key as an environment variable
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.llms.openai import OpenAI
from llama_index.embeddings.openai import OpenAIEmbedding
from google.colab import userdata

OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

# Configure LlamaIndex settings with the OpenAI API key
Settings.llm = OpenAI(api_key=OPENAI_API_KEY)
Settings.embed_model = OpenAIEmbedding(api_key=OPENAI_API_KEY)

documents = SimpleDirectoryReader("data").load_data()
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("결제금액은?")
print(response)
```
